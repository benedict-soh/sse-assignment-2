```python
remote_link = "https://github.com/sebfz1/wicket-jquery-ui"
local_link = "/mnt/e/Coding/wicket-jquery-ui"
# Uncomment to clone
# Repo.clone_from(remote_link, local_link, progress=Progress())
```

```python
repo = Repo(local_link)
fixing_commit = "3e8cfdcb0f8e6e0cf0da01e74501afb5c9bff0f"
affected_file = "WysiwygEditor.java"
file_path = "./wicket-jquery-ui-plugins/src/main/java/com/googlecode/wicket/jquery/ui/plugins/wysiwyg/"
```

# Part 3

```python
print("SSE Assignment 2 - Part 3")

commit = fixing_commit
blame_commit = ""
blame_line = -1
blamed_line = -1

while True:
    print("Performing 'git blame' on commit " + commit + "^")
    blame_fixing_commit = repo.git.blame(commit + "^", file_path + affected_file).splitlines()
    for line in blame_fixing_commit:
        print(line)
    blame_line = input("Which line introduced the problem? (Please enter a number)\n")
    blamed_line = blame_fixing_commit[int(blame_line)]
    commit = blamed_line[:10]
    user_input = input("Is this the contributing commit? [Y/N]\n")
    if user_input == "Y":
        break
        
print("VCC is " + commit)
```

# Part 5

```python
print("VCC commit information:")
show_data = repo.git.show("--stat", commit).splitlines()
for line in show_data:
    print(line)
```

```python
git_log = repo.git.log(file_path + affected_file)
authors = {}
for log_line in git_log.splitlines():
    log_info = log_line.split(" ", 1)
    if(log_info[0] == "Author:"):
        if(log_info[1] in authors):
            authors[log_info[1]] += 1
        else:
            authors[log_info[1]] = 1

print("Contributors:")
for author in authors:
    value = authors[author]
    print(author + ": " + str(value))
```
